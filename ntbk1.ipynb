# Which driver has won maximum races?

```python
# changing \\N values under 'position' column to Nan and droppping the Nan values
results = results.replace("\\N", np.nan)
results = results.dropna(subset = "position")
results
```

```python
results.info()
```

```python
# changing datatype of 'position' column from Object -> Int
results['position']=results['position'].astype(int)
```

```python
# From 'results' filtering position =1 to find the the first position in different races, and assigning it into the new variable 'driverposition'
driverposition=results[results['position']==1]
driverposition = driverposition[['raceId','driverId','position']]
driverposition
```

```python
# Fetching some columns from 'drivers' file
drivernames = drivers[['driverId','forename','surname']]
drivernames
```

```python
# Merging 'drivernames' to 'driverposition' 
driverposition = driverposition.merge(drivernames,on='driverId',how='left')
driverposition
```

```python
# Calculating total number of Wins by each driver
Wins=driverposition.groupby(['driverId','forename','surname']).sum('position')['position'].sort_values(ascending=False).reset_index(name='Wins')
```

```python
Wins
```

```python
MaxWins = Wins.head()
print("This are the first 5 rows showing the maximum wins by the drivers in different races : ")
display(MaxWins)
```

```python
# creating pie chart to display which driver has won maximum races
alt.Chart(MaxWins).mark_arc(innerRadius=15, stroke="#fff").encode(
    theta=alt.Theta("Wins", stack=True),
    radius=alt.Radius("Wins", scale=alt.Scale(type="sqrt", zero=True)),
    color=alt.Color("surname", type="ordinal"),
    tooltip=["Wins", "surname"]
).properties(
    height=400, width=400,
    title="The first 5 maxium wins by the drivers"
)
```

# Do fast pit stops contribute to winning a race?

```python
# In order to proceed with this question, took one of the 'raceId i.e 841' from 'pitstops' file to do the analysis
```

```python
pitstops_841 = pitstops[pitstops['raceId'] == 841]
pitstops_841
```

```python
# Filtering the required columns and sorting the duration in descending to check which driver spent more time during pit stops
pitstops_841=pitstops_841[['driverId','stop','duration']]
pitstops_841.sort_values(by=['duration'],ascending=False)
```

```python
# Droping null values 
results = results.replace('\\N',np.nan)
results = results.dropna()
results
```

```python
# Filtering 'race=841' from 'results' and storing the output in new variable 'race841'
race841=results[results['raceId']==841].reset_index()
race841=race841[['driverId','position']]
race841
```

```python
# Merging 'pitstops_841' to 'race841' 
race841 = race841.merge(pitstops_841,on='driverId',how='left')
```

```python
# Merging 'drivernames' to 'race841' 
race841 = race841.merge(drivernames,on='driverId',how='left')
race841
```

```python
race841=race841.sort_values('duration')
race841
```

```python
FastPitstopChart=alt.Chart(race841).mark_line().encode(
    x='position',
    y='duration',
    tooltip='surname'
    ).properties(
        width=800,
        height=500
    )
FastPitstopChart
```
